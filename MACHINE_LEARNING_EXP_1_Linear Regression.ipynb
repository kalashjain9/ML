```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import math
```

```python
#1.	Take sample values (Dataset 1)
X1 = np.array([2, 3, 4, 5, 6, 7, 8, 9, 10])
Y1 = np.array([1, 3, 6, 9, 11, 13, 15, 17, 20])

print("Dataset 1:")
print("X:", X1)
print("Y:", Y1)
```

```python
#1.	Take sample values (Dataset 1) - Second Method
data = {
    'X': [2, 3, 4, 5, 6, 7, 8, 9, 10],
    'Y': [1, 3, 6, 9, 11, 13, 15, 17, 20]
}
df = pd.DataFrame(data)
df
```

```python
# Step 1: Plot the graph between X and Y to see the trend
plt.figure(figsize=(10, 6))
plt.scatter(X1, Y1, color='blue', s=100, alpha=0.7, label='Data Points')
plt.plot(X1, Y1, color='red', linestyle='--', alpha=0.5, label='Connecting Line')
plt.xlabel('X (Independent Variable)')
plt.ylabel('Y (Dependent Variable)')
plt.title('Dataset 1: Trend Analysis (X vs Y)')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

```python
#Identifying correlation between X and Y
correlation = np.corrcoef(df['X'], df['Y'])[0,1]
print("Correlation coefficient:", correlation)
if correlation > 0.5:
    print("There is a strong positive correlation between X and Y.")
elif correlation < -0.5:
    print("There is a strong negative correlation between X and Y.")
elif -0.5 < correlation < 0.5:
    print("There is a weak correlation between X and Y.")
elif correlation == 0:
    print("There is no correlation between X and Y.")
```

```python
# Step 2: Use Linear Regression formulae to find beta0 and beta1
def calculate_regression_coefficients(x, y):
    n = len(x)

    # Calculate means
    x_mean = np.mean(x)
    y_mean = np.mean(y)

    # Calculate beta1 (slope)
    numerator = np.sum((x - x_mean) * (y - y_mean))
    denominator = np.sum((x - x_mean) ** 2)
    beta1 = numerator / denominator

    # Calculate beta0 (intercept)
    beta0 = y_mean - beta1 * x_mean

    return beta0, beta1
```

```python
# Calculate coefficients for Dataset 1
beta0, beta1 = calculate_regression_coefficients(X1, Y1)

print("Linear Regression for Dataset 1:")
print(f" (Intercept): {beta0:.4f}")
print(f" (Slope): {beta1:.4f}")
print(f"Regression Equation: Y = {beta0:.4f} + {beta1:.4f}X")

```

```python
# Calculate predictions
Y1_pred = beta0 + beta1 * X1
print("Predictions for Dataset 1:")
print(Y1_pred)
```

```python
# Calculate MSE and RMSE
n = len(X1)
mse = np.sum((Y1 - Y1_pred) ** 2)/n
rmse = np.sqrt(mse)

print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
```

```python
# Plot the regression line
plt.figure(figsize=(10, 6))
plt.scatter(X1, Y1, color='blue', s=100, alpha=0.7, label='Actual Data')
plt.scatter(X1, Y1_pred, color='green', s=100, alpha=0.7, label='Predicted Data')
plt.plot(X1, Y1_pred, color='red', linewidth=2, label=f'Regression Line: Y = {beta0:.2f} + {beta1:.2f}X')
plt.xlabel('X (Independent Variable)')
plt.ylabel('Y (Dependent Variable)')
plt.title('Dataset 1:  Linear Regression')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Step 3: Predict the value of dependent variable using the regression equation
def predict_value(x_new, beta0, beta1):
    return beta0 + beta1 * x_new

# Make predictions for new values
new_x_values = [11, 12, 15]
print("\nPredictions for Dataset 1:")
print("Using equation: Y = {:.4f} + {:.4f}X".format(beta0, beta1))
for x_new in new_x_values:
    y_pred = predict_value(x_new, beta0, beta1)
    print(f"For X = {x_new}, Predicted Y = {y_pred:.4f}")
```

```python
# Step 5: Write a program to implement Gradient Descent algorithm for Linear Regression
def gradient_descent_algorithm(X,Y,m,c,learning_rate,epochs):
  n = len(Y)
  cost = []
  for i in range(epochs):
    #Y_pred = m * X + c
    Y_pred = prediction_values(X,m,c)
    D_m = (-2/n) * sum(X * (Y - Y_pred))
    D_c = (-2/n) * sum(Y - Y_pred)
    m = m - learning_rate * D_m
    c = c - learning_rate * D_c
    cost.append(np.mean((Y - Y_pred)**2))
    if i % 100 == 0:
      print(f"Epoch {i+1}: Cost = {cost[-1]:.4f}")
      print(f"Epoch {i+1}: m = {m:.4f}, c = {c:.4f}")
  return m,c, cost
```

```python
def prediction_values(X,m,c):
  return np.array(m*X + c)

def cost_function(Y,Y_pred):
  return np.sum((Y - Y_pred)**2)/len(Y)
```

```python
# Apply Gradient Descent to Dataset 2
print("Implementing Gradient Descent for Dataset 2 (Salary Data):")
m = 0  # Initialize slope
c = 0  #Initialize intercept
iterations = 1000
cost = []
theta1, theta0, cost = gradient_descent_algorithm(df['YearsExperience'],df['Salary'],m,c,learning_rate=0.01, epochs=1000)

print(f"\nFinal Parameters:")
print(f" (Intercept): {theta0:.4f}")
print(f" (Slope): {theta1:.4f}")
print(f"Regression Equation: Salary = {theta0:.4f} + {theta1:.4f} * YearsExperience")

```

```python
#Calculating cost per epoch
plt.plot(range(iterations),cost)
plt.title("Cost per Epoch")
plt.xlabel("Epochs")
plt.ylabel("Cost")
plt.show()
```

```python
# Step 6: Predict the value of dependent variable using Gradient Descent algorithm
Y2_pred = prediction_values(df['YearsExperience'],theta1,theta0)
print("Original Salaries :",Y2)
print("Predicted Salaries :",Y2_pred)
```

```python
# Calculate MSE and RMSE for gradient descent
mse = cost_function(Y2,Y2_pred)
rmse = np.sqrt(mse)
print(f"\nGradient Descent Performance:")
print(f"MSE: {mse:.2f}")
print(f"RMSE: {rmse:.2f}")
```

```python
# Make predictions
Y2_pred_sklearn = sklearn_model.predict(X2_reshaped)
print("Original Salaries :",Y2)
print("Predicted Salaries :",Y2_pred_sklearn)
```

```python
# Calculate MSE and RMSE
mse_sklearn = mean_squared_error(Y2, Y2_pred_sklearn)
rmse_sklearn = np.sqrt(mse_sklearn)

print(f"\nScikit-learn Performance:")
print(f"MSE: {mse_sklearn:.2f}")
print(f"RMSE: {rmse_sklearn:.2f}")
print(f"R Score: {sklearn_model.score(X2_reshaped, Y2):.4f}")

```
